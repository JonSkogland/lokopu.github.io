My first go at using the dashboard format in Quarto.
This is some analytics on Uber bookings. Retrieved from Kaggle


```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data/day26/ncr_ride_bookings.csv')
```

```python
grouped = df.groupby(['Vehicle Type', 'Date'], as_index=False).agg(bookings=('Booking ID', 'count'), Booking_value=('Booking Value', 'sum'))
```

## Row {height=50%}

```python
px.scatter(
    grouped, x='Date', y='bookings',
    color='Vehicle Type'
)
```
